```
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=3, input_shape=[1])])

model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

model.fit(xs, ys, epochs=500)

```

```
print(model.predict([5.0]))
```

```
from math import sin, sqrt, pow

import numpy as np
from tensorflow import keras

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf


def funkcja(x, y):
  return sin(sqrt(pow(x, 2) + pow(y, 2)))/sqrt(pow(x, 2) + pow(y, 2))

xs_ = np.arange(-20.0, 20.1, 0.1, dtype = float)
ys_ = np.arange(-20.0, 20.1, 0.1, dtype = float)
zs = []
xy = []
for x,y in zip(xs_, ys_):
  zs.append(funkcja(x, y))
  xy.append([x, y])
xy = np.asarray(xy, dtype = float)
zs = np.asarray(zs, dtype = float)
model_ = tf.keras.Sequential([
                             keras.layers.Dense(1000, input_shape = [2]),
                             keras.layers.Dense(1000),
                             keras.layers.Dense(1000),
                             keras.layers.Dense(1000),
                             keras.layers.Dense(1000),
                             keras.layers.Dense(1)
                             ])
model_.compile(optimizer = 'adam', loss = 'mean_squared_error')
model_.fit(xy, zs, epochs = 200)
```

```
print(model_.predict([[5.0, 5.0]]))
```
