```
import tensorflow as tf
import numpy as np
from tensorflow import keras


def f(x):
    return (x*x)/2
#Metoda Trapezow
def mTrapezow(a, b , n):
    h = ((b-a) /n)
    sum = f(a)/2+f(b)/2

    for i in range(1, n):
        sum += f(a+(i/n) * (b-a))

    return h*sum
#Metoda Simpsona
def mSimpsona(a,b,n):
    sum = f(a) + f(b)
    h = 0.0
    array = []

    array.append(a)
    
    # xi
    for i in range(1,n):
        array.append(a + (i / n) * (b - a))
        sum+=2 * f(a+(i / n) * (b-a))

    array.append(b)

    # ti
    for i in range(0, n):
        sum += 4 * f((array[i + 1] + array[i]) / 2)
        h = (array[i + 1] - array[i]) / 2

    return sum * (h / 3)



print('mTrapezow ', mTrapezow(0.5, 1.5, 20))
print('mSimpsona ', mSimpsona(0.5, 1.5, 20))
```
